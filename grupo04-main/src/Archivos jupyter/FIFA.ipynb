```python
import csv
import os

dir = os.path.realpath(".")
dir_entrada = os.path.join(dir,"FIFA-21 Complete.csv")
dir_salida = os.path.join(dir,"FIFA-21 Complete Modificado.csv")

with open(dir_entrada,"r",encoding="utf8") as fifa_file:
    
    fifa_reader = csv.reader(fifa_file, delimiter=';')
    header = next(fifa_reader)    
    
    with open(dir_salida,"w",encoding="utf8") as present_file:
        
        present_writer = csv.writer(present_file, delimiter=';')
        lineas = []

        for row in fifa_reader:
            
            if (int(row[7]) < 60):
                row[7] = "Regular"
                
            elif (60 <= int(row[7]) <= 79):
                row[7] = "Bueno"
                
            elif (80 <= int(row[7]) <= 89):
                row[7] = "Muy bueno"
                
            else:
                row[7] = "Sobresaliente"
                
            row[3] = row[3].replace("CDM", "Mediocampista Defensivo|Mediocampista Central")
            row[3] = row[3].replace("CDM", "Mediocampista Defensivo") 
            row[3] = row[3].replace("GK", "Arquero")
            row[3] = row[3].replace("SW", "Libero")
            row[3] = row[3].replace("LB", "Defensor Izquierdo")
            row[3] = row[3].replace("RB", "Defensor Derecho")
            row[3] = row[3].replace("LWB", "Defensor Izquierdo")
            row[3] = row[3].replace("RWB", "Defensor Derecho")
            row[3] = row[3].replace("CB", "Defensor Central")
            row[3] = row[3].replace("LCB", "Defensor Central")
            row[3] = row[3].replace("RCB", "Defensor Central")
            row[3] = row[3].replace("DM", "Mediocampista Defensivo")
            row[3] = row[3].replace("LDM", "Mediocampista Defensivo")
            row[3] = row[3].replace("RDM", "Mediocampista Defensivo")
            row[3] = row[3].replace("CM", "Mediocampista Central")
            row[3] = row[3].replace("LCM", "Mediocampista Central")
            row[3] = row[3].replace("RCM", "Mediocampista Central")
            row[3] = row[3].replace("LM", "Mediocampista Izquierdo")
            row[3] = row[3].replace("RM", "Mediocampista Derecho")
            row[3] = row[3].replace("CAM", "Mediocampista Ofensivo")
            row[3] = row[3].replace("LAM", "Mediocampista Ofensivo")
            row[3] = row[3].replace("RAM", "Mediocampista Ofensivo")
            row[3] = row[3].replace("SS", "Segundo Delantero")
            row[3] = row[3].replace("LW", "Delantero Izquierdo")
            row[3] = row[3].replace("RW", "Delantero Derecho")
            row[3] = row[3].replace("CF", "Delantero Central")
            row[3] = row[3].replace("LF", "Delantero Central")
            row[3] = row[3].replace("RF", "Delantero Central")
            row[3] = row[3].replace("ST", "Delantero Central")
            
            lineas.append(row)
            lineas_para_tarjetas = map(lambda row: [row[8], row[2], row[3], row[5], row[7], row[1]], lineas)
    
        present_writer.writerows(lineas_para_tarjetas)




```
